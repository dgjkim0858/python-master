코드 14-1

```python
import pandas as pd

df = pd.read_csv('diabetes.csv')
```

코드 14-2

```python
print(len(df))
display(df.head())
```

코드 14-3

```python
import numpy as np

columns = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
for col in columns:
  df[col].replace(0, np.nan, inplace=True)
```

코드 14-4

```python
df.dropna(inplace=True)
print(len(df))
```

코드 14-5

```python
target = df.pop('Outcome')
```

코드 14-6

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.2,
                                                    random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

코드 14-7

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

코드 14-8

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(12, activation='relu', input_dim=8))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

코드 14-9

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

코드 14-10

```python
history = model.fit(X_train, y_train, epochs=30, batch_size=5,
                    validation_data=(X_test, y_test))
```

코드 14-11

```python
print(model.evaluate(X_train, y_train))
print(model.evaluate(X_test, y_test))
```

코드 14-12

```python
import matplotlib.pyplot as plt

epochs = np.arange(1, len(history.history['loss']) + 1)

plt.figure(figsize=(5,3))
plt.plot(epochs, history.history['loss'], label='Training loss')
plt.plot(epochs, history.history['val_loss'], label='Validation loss')
plt.plot(epochs, history.history['accuracy'], label='Training acc')
plt.plot(epochs, history.history['val_accuracy'], label='Validation acc')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

LAB 붓꽃 품종 분류하기

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt

# 데이터 수집하기
iris = datasets.load_iris()
print(iris['data'][:5])
print(iris['target'][:5])

# 데이터 전처리하기
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'],
                                                    random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

# 모델 생성하기
model = Sequential()
#model.add(Dense(64, activation='relu', input_shape=(4,)))
model.add(Dense(64, activation='relu', input_dim=4))
model.add(Dense(64, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test))

# 모델 평가하기
print(model.evaluate(X_train, y_train))
print(model.evaluate(X_test, y_test))

epochs = np.arange(1, len(history.history['loss']) + 1)
plt.figure(figsize=(7,4))
plt.plot(epochs, history.history['loss'], label='Training loss')
plt.plot(epochs, history.history['val_loss'], label='Validation loss')
plt.plot(epochs, history.history['accuracy'], label='Training acc')
plt.plot(epochs, history.history['val_accuracy'], label='Validation acc')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

LAB 자동차 연비 예측하기

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt

# 데이터 수집하기
mpg = sns.load_dataset('mpg')
display(mpg.head())

# 데이터 전처리하기
mpg.isnull().sum()
mpg = mpg.dropna()
target = mpg.pop('mpg')
data = mpg.drop(columns=['origin', 'name'])
display(data.head())

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 모델 생성하기
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=6))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
history = model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))

# 모델 평가하기
print(model.evaluate(X_train, y_train))
print(model.evaluate(X_test, y_test))

epochs = np.arange(1, len(history.history['loss']) + 1)
plt.figure(figsize=(5,3))
plt.plot(epochs, history.history['loss'], label='Training loss')
plt.plot(epochs, history.history['val_loss'], label='Validation loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()

y_pred = model.predict(X_test)
ind = np.arange(len(y_pred))
plt.figure(figsize=(6,3))
plt.plot(ind, y_pred, label='pred')
plt.plot(ind, y_test, label='real')
plt.ylabel('MPG')
plt.legend()
plt.show()
```

**[보충] 패션 MNIST 분류하기**

```python
from tensorflow.keras.datasets import fashion_mnist
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(3,3))
plt.imshow(X_train[0])
plt.show()
```

```python
print(y_train[0])
```

```python
plt.figure(figsize=(5,5))
for i in range(9):
  plt.subplot(3, 3, i+1)
  plt.imshow(X_train[i], cmap='binary')
  plt.axis('off')
plt.show()
```

```python
X_train, X_test = X_train/255.0, X_test/255.0
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense

model = Sequential()
model.add(Flatten(input_shape=[28, 28]))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=10, validation_split=0.2)
```

```python
print(model.evaluate(X_train, y_train))
print(model.evaluate(X_test, y_test))
```

```python
import numpy as np
import matplotlib.pyplot as plt

epochs = np.arange(1, len(history.history['loss']) + 1)

plt.figure(figsize=(5,3))
plt.plot(epochs, history.history['loss'], label='Training loss')
plt.plot(epochs, history.history['val_loss'], label='Validation loss')
plt.plot(epochs, history.history['accuracy'], label='Training acc')
plt.plot(epochs, history.history['val_accuracy'], label='Validation acc')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```
