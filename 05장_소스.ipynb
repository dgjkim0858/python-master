순번을 입력받아 왕 묘호 출력하기

```python
dynasties = ['태조','정종','태종','세종','문종','단종','세조','예종','성종',
             '연산군','중종','인종','명종','선조','광해군','인조','효종',
             '현종','숙종','경종','영조','정조','순조','헌종','철종', '고종', '순종']

answer = input('순번 또는 왕 이름: ')
if answer.isdigit():
  num = int(answer)
  if num>=1 and num<=27:
    print(dynasties[num-1])
  else:
    print('1~27대까지 있습니다.')
else:
  if answer in dynasties:
    print(dynasties.index(answer)+1, '대 왕입니다.')
  else:
    print(answer, '이름을 가진 왕은 없습니다.')
```

**[보충] 상식 퀴즈**

상식 퀴즈 프로그램

```python
problems = ['우리나라 국보 1호는?','우리나라 보물 1호는?','우리나라에서 가장 높은 산은?',
           '우리나라에서 가장 긴 강은?','세계에서 가장 높은 산은?','세계에서 가장 긴 강은?']
answers= ['숭례문','흥인지문','한라산','낙동강','에베레스트','아마존']

for i in range(len(problems)):
  ans = input(problems[i])
  if ans == answers[i]:
    print('맞았습니다.')
  else:
    print('틀렸습니다.')
```

업그레이드 상식 퀴즈 프로그램

```python
import random

problems = ['우리나라 국보 1호는?','우리나라 보물 1호는?','우리나라에서 가장 높은 산은?',
           '우리나라에서 가장 긴 강은?','세계에서 가장 높은 산은?','세계에서 가장 긴 강은?']
answers= ['숭례문','흥인지문','한라산','낙동강','에베레스트','아마존']

num = int(input('몇 문제 풀어보겠습니까(1~6)? '))

for i in range(num):
  idx = random.randint(0,len(problems)-1)
  ans = input(problems[idx])
  if ans == answers[idx]:
    print('맞았습니다.')
  else:
    print('틀렸습니다.')
  del problems[idx], answers[idx]
```
