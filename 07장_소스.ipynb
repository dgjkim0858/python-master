LAB 주문 금액 계산기

```python
prices = {'아메리카노':2000,'라떼':3000,'녹차':2500}
sum = 0

while True:
  s = input('음료 종류 또는 종료: ')
  if s=='종료':
    break
  elif s in prices:
    n = int(input(f'{s} 주문 수: '))
    sum += prices[s]*n
  else:
    print(f'{s}는 없습니다.')

print(f'총 금액: {sum}원')
```

LAB 혈액형 빈도수

```python
blood = {'A':0,'B':0,'O':0,'AB':0}

while True:
  s = input('혈액형(A, B, O, AB) 또는 종료: ')
  if s=='종료':
    print('종료합니다.')
    break
  elif s in blood:
    blood[s] += 1
  else:
    print('잘못 입력했습니다.')

for key, value in blood.items():
  print(key, value)
```

**[플러스] 프로그래밍 언어 선호도 조사**

선호하는 프로그래밍 언어 조사

```python
langs = {'Python':0,'C':0,'Java':0,'JavaScript':0}

while True:
  s = input('선호하는 언어(Python,C,Java,JavaScript) 또는 종료: ')
  if s=='종료':
    print('종료합니다.')
    break
  elif s in langs:
    langs[s] += 1
  else:
    print(f'{s}는 등록되지 않은 언어입니다.')

for key, value in langs.items():
  print(key, value)
```

사용 가능한 프로그래밍 언어 조사

```python
langs = {}

while True:
  s = input('사용 가능한 언어(,로 구분) 또는 종료: ')
  s = s.lower()
  if s=='종료':
    print('종료합니다.')
    break
  for a in s.split(','):
    if a in langs:
      langs[a] += 1
    else:
      langs[a] = 1

for key, value in langs.items():
  print(key, value)
```

**[플러스] 영어 단어 퀴즈**

동물 이름 퀴즈

```python
words = {'사자':'lion','호랑이':'tiger','토끼':'rabbit'}

for key, value in words.items():
  ans = input(key+' 영어 이름은? ')
  if ans == value:
    print('맞았습니다.')
  else:
    print('틀렸습니다.')
```

영어 단어 퀴즈

```python
import googletrans
translator = googletrans.Translator()

words = {}

def registration():
  s = input('등록할 단어: ')
  if s in words:
    print('등록된 단어입니다.')
  else:
    words[s] = translator.translate(s, dest='en').text

def quiz():
  for key, value in words.items():
    ans = input(key+' 영어 이름은? ')
    if ans == value:
      print('맞았습니다.')
    else:
      print('틀렸습니다.')

while True:
  sel = int(input('1.등록 2.퀴즈 3.종료: '))
  if sel==3:
    print('종료합니다.')
    break
  elif sel==1:
    registration()
  elif sel==2:
    quiz()
  else:
    print('잘못 입력했습니다.')
```

환전 계산기

```python
from currency_converter import CurrencyConverter

cc = CurrencyConverter('http://www.ecb.europa.eu/stats/eurofxref/eurofxref.zip')
curr_code = {'대한민국':'KRW','미국':'USD','유럽연합':'EUR','일본':'JPY'}

while True:
  start = input('시작 통화 국가명 또는 종료: ')
  if start=='종료':
    break
  target = input('목표 통화 국가명: ')
  money = int(input('시작 통화 금액: '))
  if start in curr_code and target in curr_code:
    print(cc.convert(money,curr_code[start],curr_code[target]))
  else:
    if start not in curr_code:
      print(f'{start}이 등록되지 않았습니다.')
    if target not in curr_code:
      print(f'{target}이 등록되지 않았습니다.')
```
