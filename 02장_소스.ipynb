**[플러스] 시간 단위 변환하기**

분 단위 시간을 시간과 분 단위로 변환하기

```python
min = int(input('분 단위 시간: '))
hour = min // 60
min = min % 60
print(f'{hour}시간 {min}분')
```

초 단위 시간을 시간, 분, 초 단위로 변환하기

```python
second = int(input('초 단위 시간: '))
hour = second // 3600
second = second % 3600
min = second // 60
second = second % 60
print(f'{hour}시간 {min}분 {second}초')
```

**[플러스] 주문 금액 계산기**

주문 금액 계산기 (1)

```python
americano = int(input('아메리카노 주문 수: '))
latte = int(input('라떼 주문 수: '))
greentea = int(input('녹차 주문 수: '))

sum = americano*2000 + latte*3000 + greentea*2500

print('주문 금액:', sum)
```

주문 금액 계산기 (2)

```python
americano_price = 2000
latte_price = 3000
greentea_price = 2500

sum = 0

americano_num = int(input('아메리카노 주문 수: '))
sum += americano_num * americano_price

latte_num = int(input('라떼 주문 수: '))
sum += latte_num * latte_price

greentea_num = int(input('녹차 주문 수: '))
sum += greentea_num * greentea_price

print('총 금액:', sum)
```

**[보충] 연비 계산기**

연비 계산하기

```python
distance = int(input('주행거리(km): '))
fuel_amount = int(input('사용연료량(l): '))
fuel_efficiency = distance / fuel_amount
print(f'연비: {fuel_efficiency:.2f}km/l')
```

주행 가능한 거리 계산하기

```python
fuel_efficiency = float(input('연비(km/l): '))
fuel_amount = int(input('연료량(l): '))
distance = fuel_efficiency * fuel_amount
print(f'주행 가능 거리: {distance:.2f}km')
```
