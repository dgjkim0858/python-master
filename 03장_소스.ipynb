LAB 함수를 이용해서 사칙연산하기

```python
def add(x, y):
  return x+y

def sub(x, y):
  return x-y

def mul(x, y):
  return x*y

def div(x, y):
  return x/y

a = int(input('정수: '))
b = int(input('정수: '))
print(f'{a}+{b}={add(a,b)}')
print(f'{a}-{b}={sub(a,b)}')
print(f'{a}*{b}={mul(a,b)}')
print(f'{a}/{b}={div(a,b)}')
```

LAB 함수 하나로 몫과 나머지 반환하기

```python
def div(x, y):
  return x//y, x%y

a = int(input('정수: '))
b = int(input('정수: '))
quotient, remainder = div(a, b)
print(f'{a}를 {b}로 나눈 몫은 {quotient}이고 나머지는 {remainder}')
```

**[플러스] 두 도형의 넓이 비교하기**

```python
import math

def square():
  width = int(input('사각형의 가로: '))
  height = int(input('사각형의 세로: '))
  return width*height

def triangle():
  base = int(input('삼각형의 밑변: '))
  height = int(input('삼각형의 높이: '))
  return base*height/2

def circle():
  radius = int(input('원의 반지름: '))
  return math.pi*radius*radius

shape_type1 = input('도형의 종류(사각형, 삼각형, 원): ')
if shape_type1 == '사각형':
  area1 = square()
elif shape_type1 == '삼각형':
  area1 = triangle()
else:
  area1 = circle()

shape_type2 = input('도형의 종류(사각형, 삼각형, 원): ')
if shape_type2 == '사각형':
  area2 = square()
elif shape_type2 == '삼각형':
  area2 = triangle()
else:
  area2 = circle()

if area1 > area2:
  print(f'도형1(넓이:{area1})이 도형2(넓이:{area2})보다 {area1-area2}만큼 넓습니다.')
elif area1 < area2:
  print(f'도형2(넓이:{area2})가 도형1(넓이:{area1})보다 {area2-area1}만큼 넓습니다.')
else:
  print(f'두 도형의 넓이는 {area1}로 같습니다.')
```

**[보충] 지하철 운임 계산기**

이동 거리 구간에 따른 운임 체계

```python
distance = int(input('승차 거리: '))
fare = ((distance+4)//5)*100
print(fare)
```

기본 운임이 있는 운임 체계

```python
fare = 1250
distance = int(input('승차 거리: '))
if distance>10:
  fare += ((distance-10+4)//5)*100
print(fare)
```

지하철 운임 계산기

```python
base_fare = 1250
distance = int(input('승차 거리: '))

if distance <= 10:
  fare = base_fare
elif distance <= 50:
  fare = base_fare + ((distance-10+4)//5)*100
else:
  fare = base_fare + 8*100 + ((distance-50+7)//8)*100
print(f'요금: {fare}')
```
