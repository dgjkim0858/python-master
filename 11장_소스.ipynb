```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('지역별면적.csv', encoding='EUC-KR')
display(df1.head())

df2 = pd.read_csv('지역별인구수.csv', encoding='EUC-KR')
display(df2.head())

df = pd.merge(df1, df2, on='지역')
display(df.head())

df['인구밀도'] = df['총인구수'] / df['면적']
display(df.head())

display(df[df['인구밀도']>=2000])

ind = np.arange(len(df))
plt.figure(figsize=(5,3))
plt.bar(ind, df['인구밀도'])
plt.xlabel('지역', fontproperties=font11)
plt.ylabel('인구밀도', fontproperties=font11)
plt.xticks(ind, df['지역'], rotation='vertical', fontproperties=font9)
plt.show()
```

**[플러스] 자동차 배기량과 CO2 배출량의 관계 분석하기**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('car20240731.csv')
print(len(df))
display(df.head())
```

```python
display(df.describe())
```

```python
df = df[df['배기량']!=0]
print(len(df))
display(df.describe())
```

```python
df1 = df[df['연료']=='휘발유']
df2 = df[df['연료']=='경유']
```

```python
plt.figure(figsize=(5,3))
plt.scatter(df1['배기량'], df1['CO2배출량(g_km)'], s=3, alpha=0.5, label='Gasoline')
plt.scatter(df2['배기량'], df2['CO2배출량(g_km)'], s=3, alpha=0.5, label='Diesel')
plt.legend()
plt.xlabel('displacement')
plt.ylabel('CO2 emissions')
plt.show()
```

```python
cor = df1['배기량'].corr(df1['CO2배출량(g_km)'])
print(cor)

cor = df2['배기량'].corr(df2['CO2배출량(g_km)'])
print(cor)
```

**[보충] 자동차 중량과 연비의 산점도 그리기**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'모델명': ['아반떼', '셀토스', '토레스'],
                   '중량': ['1,265kg', '1,405kg', '1,520kg'],
                   '연비': ['15.3km/l', '12.8km/l', '11.2km/l']})

display(df)

df['중량'] = df['중량'].str.rstrip('kg').str.replace(',','').astype(int)
df['연비'] = df['연비'].str.rstrip('km/l').astype(float)
display(df)

plt.figure(figsize=(5,3))
plt.scatter(df['중량'], df['연비'])
plt.xlim(1200,1600)
plt.ylim(10,16)
plt.xlabel('weight(kg)')
plt.ylabel('fuel efficiency(km/l)')
plt.show()
```
