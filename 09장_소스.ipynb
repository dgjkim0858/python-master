LAB 코로나19 확진자 수 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

months = np.arange(1, 13)
num_cases = [11, 2920, 6636, 750, 615, 1141, 1044, 5473, 3707, 2444, 7324, 26198]
cum_cases = np.cumsum(num_cases)

plt.figure(figsize=(5, 3))
plt.plot(months, num_cases, label='number of cases')
plt.plot(months, cum_cases, label='accumulated number of cases')
plt.xlabel('month')
plt.legend()
plt.show()
```

LAB 동전 던지기 시뮬레이션

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 3))
for i in range(3):
  coin = np.random.randint(0, 2, 5000)
  ind = np.arange(1, len(coin)+1)
  p_c = np.cumsum(coin)/ind*100
  plt.plot(ind, p_c)
plt.ylim(0,100)
plt.ylabel('%')
plt.show()
```

**[플러스] 출생아 수 시각화**

남아 출생아 수와 여아 출생아 수 막대 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

year = [2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
boy = [248958,223883,223356,224906,208064,184308,167686,155416,139362,133516]
girl = [235592,212572,212079,213514,198179,173463,159136,147260,132975,127046]

ind = np.arange(len(year))

w = 0.4
plt.figure(figsize=(6, 3.5))
plt.bar(ind-w/2, boy, color='b', width=w, label='men')
plt.bar(ind+w/2, girl, color='r', width=w, label='women')
plt.xticks(ind, year)
plt.legend()
plt.xlabel('year')
plt.ylabel('number of births')
plt.show()
```

출생아 수 누적 막대 그래프 그리기

**[보충] 우리나라 대통령 키 분석하기**

```python
import numpy as np
import matplotlib.pyplot as plt

name = ['이승만', '윤보선', '박정희', '최규하', '전두환', '노태우',
        '김영삼', '김대중', '노무현', '이명박', '박근혜', '문재인', '윤석열']
height = np.array([170, 169, 164, 178, 171, 178, 168, 173, 168, 173, 162, 172, 178])

print('최대:', np.max(height))
print('최소:', np.min(height))
print('평균:', np.mean(height))
print('표준편차:', np.std(height))

plt.figure(figsize=(5, 3))
plt.hist(height, edgecolor='#FFFFFF', color='#EE6666')
plt.xlabel('height')
plt.ylabel('number')
plt.show()
```

**[보충] 기초대사량 계산기**

기초대사량 구하기

```python
gender = input('성별(남자 또는 여자): ')
height = float(input('키: '))
weight = int(input('몸무게: '))
age = int(input('나이: '))

if gender == '남자':
  bmr = 66.47 + (13.75 * weight) + (5 * height) - (6.76 * age)
else:
  bmr = 665.1 + (9.56 * weight) + (1.85 * height) - (4.68 * age)

print('당신의 기초대사량:', bmr)
```

나이에 따른 기초대사량 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

gender = input('성별(남자 또는 여자): ')
height = float(input('키: '))
weight = int(input('몸무게: '))

age = np.linspace(20, 80, 7)

if gender == '남자':
  bmr = 66.47 + (13.75 * weight) + (5 * height) - (6.76 * age)
else:
  bmr = 665.1 + (9.56 * weight) + (1.85 * height) - (4.68 * age)

plt.figure(figsize=(5, 3))
plt.plot(age, bmr)
plt.xlabel('age')
plt.ylabel('basal metabolism')
plt.show()
```

**[보충] 만유인력 계산기**

만유인력 구하기

```python
m1 = int(input('물체1 질량(kg): '))
m2 = int(input('물체2 질량(kg): '))
r = int(input('거리(m): '))
G = 6.674e-11
F = G*m1*m2/(r**2)
print('만유인력:', F)
```

거리에 따른 만유인력 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

m1 = int(input('물체1 질량(kg): '))
m2 = int(input('물체2 질량(kg): '))
G = 6.674e-11
r = np.arange(100, 1001, 10)
F = G*m1*m2/(r**2)

plt.figure(figsize=(4.5, 3))
plt.plot(r, F)
plt.xlabel('distance(m)')
plt.ylabel('gravitational force in newtons')
plt.show()
```
