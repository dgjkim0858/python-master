```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import auth
auth.authenticate_user()
import gspread
from oauth2client.client import GoogleCredentials
from google.auth import default
creds, _ = default()
gc = gspread.authorize(creds)
url = 'https://docs.google.com/spreadsheets/d/1BTkjB3U349zppIunNiaWy-2fS-W7-SzLHOD91rblG9c/edit?gid=790703349#gid=790703349'   # 복사한 스프레드시트 URL 붙이기
worksheet = gc.open_by_url(url).sheet1
rows = worksheet.get_all_values()
df = pd.DataFrame.from_records(rows[1:], columns=rows[0])

display(df)
```

```python
df['키'] = df['키'].astype(int)
df['몸무게'] = df['몸무게'].astype(int)

plt.figure(figsize=(5,3))
plt.scatter(df['키'], df['몸무게'])
for _, row in df.iterrows():
  plt.annotate(row['이름'], xy=(row['키'],row['몸무게']), xytext=(5,-10), \
  textcoords='offset points', fontproperties=font9)
plt.xlim(163,178)
plt.ylim(43,95)
plt.xlabel('키', fontproperties=font11)
plt.ylabel('몸무게', fontproperties=font11)
plt.show()

df['체질량지수'] = df['몸무게'] / ((df['키']/100.0)*(df['키']/100.0))

df['판정'] = np.where(df['체질량지수']<18.5, '저체중',
                   np.where(df['체질량지수']<23, '정상',
                      np.where(df['체질량지수']<25, '과체중',
                          np.where(df['체질량지수']<30, '1단계비만',
                              np.where(df['체질량지수']<35, '2단계비만', '고도비만')))))

display(df)

result = {'저체중':0,'정상':0,'과체중':0,'1단계비만':0,'2단계비만':0,'고도비만':0}

for x in df['판정']:
  result[x] = result[x] + 1

plt.figure(figsize=(5,3))
ind = np.arange(len(result))
plt.bar(ind, result.values(), width=0.7)
plt.xticks(ind, result.keys(), fontproperties=font9)
plt.show()
```

```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent='Jeju v1.0')
center = geolocator.geocode('서울특별시')

df = pd.DataFrame()
df['이름'] = ['경복궁', '덕수궁', '창경궁', '창덕궁']

lat = []
lng = []
for _, row in df.iterrows():
  loc = geolocator.geocode(row['이름'])
  lat.append(loc.latitude)
  lng.append(loc.longitude)

df['위도'] = lat
df['경도'] = lng

m = folium.Map(location=[center.latitude, center.longitude],
               zoom_start=12,
               width=640,
               height=480
)

for _, row in df.iterrows():
  folium.Marker([row['위도'], row['경도']], tooltip=row['이름']).add_to(m)

m
```

**[보충] 프로그래밍 언어 선호도 설문 분석하기**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import auth
auth.authenticate_user()
import gspread
from oauth2client.client import GoogleCredentials
from google.auth import default
creds, _ = default()
gc = gspread.authorize(creds)
url = '스프레드시트 URL'   # 복사한 스프레드시트 URL 붙이기
worksheet = gc.open_by_url(url).sheet1
rows = worksheet.get_all_values()
df = pd.DataFrame.from_records(rows[1:], columns=rows[0])

display(df.head())
```

```python
result = {'Python':0,'C':0,'Java':0,'C++':0,'C#':0,'JavaScript':0}
```

```python
for x in df['선호하는 프로그래밍 언어(복수 가능)']:
  for a in x.split(', '):
    result[a] += 1
```

```python
for key, value in result.items():
  print(key, value)
```

```python
ind = np.arange(len(result))
plt.figure(figsize=(5,3))
plt.bar(ind, result.values(), width=0.5)
plt.xticks(ind, result.keys(), rotation='vertical')
plt.show()
```

```python
result_men = {'Python':0,'C':0,'Java':0,'C++':0,'C#':0,'JavaScript':0}
for x in df[df['성별']=='남자']['선호하는 프로그래밍 언어(복수 가능)']:
  for a in x.split(', '):
    result_men[a] += 1
```

```python
result_women = {'Python':0,'C':0,'Java':0,'C++':0,'C#':0,'JavaScript':0}
for x in df[df['성별']=='여자']['선호하는 프로그래밍 언어(복수 가능)']:
  for a in x.split(', '):
    result_women[a] += 1
```

```python
ind = np.arange(len(result_men))
plt.figure(figsize=(5,3))
plt.bar(ind-0.2, result_men.values(), width=0.4, label='men')
plt.bar(ind+0.2, result_women.values(), width=0.4, label='women')
plt.xticks(ind, result_men.keys(), rotation='vertical')
plt.legend()
plt.show()
```
