[LAB] 와인 분류하기

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# 데이터 수집하기
wine = datasets.load_wine()
print('wine 키:', wine.keys())
print('데이터:\n', wine['data'][:5])
print('특성 이름:', wine['feature_names'])
print('타깃:\n', wine['target'][:5])
print('타깃 이름:', wine['target_names'])
print('데이터 크기:', wine['data'].shape)
print('타깃 크기:', wine['target'].shape)

# 데이터 전처리하기
X_train, X_test, y_train, y_test = train_test_split(wine['data'], wine['target'],
                                                    random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

# 모델 생성하기
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# 모델 이용해 예측하기
y_pred = knn.predict(X_test)
print('테스트 세트 예측 값:\n', y_pred)
print('테스트 세트 실제 타깃:\n', y_test)

# 모델 평가하기
print(f'훈련 세트 정확도: {knn.score(X_train, y_train):.3f}')
print(f'테스트 세트 정확도: {knn.score(X_test, y_test):.3f}')
```

[LAB] 당뇨병 진행 정도 예측하기

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np

# 데이터 수집하기
diabetes = datasets.load_diabetes()
print('diabetes 키:', diabetes.keys())
print('데이터:\n', diabetes['data'][:3])
print('특성 이름:', diabetes['feature_names'])
print('타깃:\n', diabetes['target'][:3])
print('데이터 크기:', diabetes['data'].shape)
print('타깃 크기:', diabetes['target'].shape)

# 데이터 전처리하기
X_train, X_test, y_train, y_test = train_test_split(diabetes['data'], diabetes['target'],
                                   random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

# 모델 생성하기
lr = LinearRegression()
lr.fit(X_train, y_train)

print('절편:', np.round(lr.intercept_,3))
print('계수:', np.round(lr.coef_,3))

# 모델 이용해 예측하기
pred = lr.predict(X_test)
print('테스트 세트 예측 값:\n', np.round(pred[:10],3))
print('테스트 세트 실제 타깃:\n', y_test[:10])

# 모델 평가하기
print(f'훈련 세트 정확도: {lr.score(X_train, y_train):.3f}')
print(f'테스트 세트 정확도: {lr.score(X_test, y_test):.3f}')
```

**[플러스] MNIST 분류하기**

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', as_frame=False, parser='auto')
```

```python
print(mnist.keys())
```

```python
print('데이터 크기:', mnist['data'].shape)
print('타깃 크기:', mnist['target'].shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(2,2))
img = mnist['data'][0].reshape(28,28)
plt.imshow(img, cmap='binary')
plt.show()
```

```python
print(mnist['target'][0])
```

```python
plt.figure(figsize=(4,4))
for i in range(9):
  plt.subplot(3,3,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(mnist['data'][i].reshape(28,28), cmap='binary')
  plt.xlabel(mnist['target'][i])
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(mnist['data'], mnist['target'], random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
print('테스트 세트 예측 값:\n', y_pred[:10])
print('테스트 세트 실제 타깃:\n', y_test[:10])
```

```python
print(f'훈련 세트 정확도: {knn.score(X_train, y_train):.3f}')
print(f'테스트 세트 정확도: {knn.score(X_test, y_test):.3f}')
```

**[보충] 자동차 연비 예측하기**

```python
import seaborn as sns

mpg = sns.load_dataset('mpg')
```

```python
display(mpg.head())
```

```python
mpg.isnull().sum()
```

```python
mpg = mpg.dropna()
```

```python
target = mpg.pop('mpg')
```

```python
data = mpg.drop(columns=['origin', 'name'])
display(data.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
import numpy as np

print('절편:', np.round(lr.intercept_,3))
print('계수:', np.round(lr.coef_,3))
```

```python
y_pred = lr.predict(X_test)
print('테스트 세트 예측 값:\n', np.round(y_pred[:10],3))
print('테스트 세트 실제 타깃:\n', list(y_test[:10]))
```

```python
import numpy as np
import matplotlib.pyplot as plt

ind = np.arange(len(X_test))
plt.figure(figsize=(8,3))
plt.plot(ind, y_test, label='real')
plt.plot(ind, y_pred, label='pred')
plt.ylabel('mpg')
plt.legend()
plt.show()
```

```python
print(f'훈련 세트 정확도: {lr.score(X_train, y_train):.3f}')
print(f'테스트 세트 정확도: {lr.score(X_test, y_test):.3f}')
```
